```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Task 2 Part 1


```python
# loading csv file
df_part_1 = pd.read_csv('task2-1.csv')
df_part_1.head()
```

```python
# Creating a new column by combining 'business' and 'Month' columns
df_part_1['business_month'] = df_part_1['business'] + '-' + df_part_1['Month'].astype(str)

# Scaling the trip_count values using logarithmic scaling
df_part_1['log_trip_count'] = np.log10(df_part_1['trip_count'])

# Plotting using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=df_part_1, x='business_month', y='log_trip_count', hue='business')
plt.xlabel('Business-Month')
plt.ylabel('Log(Trip Count)')
plt.title('Trip Count by Business and Month (Logarithmic Scale)')
plt.legend()
plt.show()
```
