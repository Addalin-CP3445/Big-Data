```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# loading csv file
df = pd.read_csv('task5.csv')
df.head()
```

```python
# Plotting using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='day', y='average_waiting_time')
plt.xlabel('Day in January')
plt.ylabel('Average Waiting Time (seconds)')
plt.title('Average Waiting Time for Month of January')
plt.show()
```
